```python
from src.infrastructure import settings
import requests
import json
```

```python
filename = "src/infrastructure/api/prompt_headers/question.md"
with open(filename) as f:
    question_header = f.read()

qeustion = "Американец Джон Форд - единственный в истории обладатель четырёх этих премий за лучшую режиссуру"
content = question_header + qeustion
print(content)
```

```python
models = [
    # "xiaomi/mimo-v2-flash:free",
    # "tngtech/deepseek-r1t2-chimera:free",
    # "tngtech/deepseek-r1t-chimera:free",
    # "z-ai/glm-4.5-air:free",
    "nvidia/nemotron-3-nano-30b-a3b:free",
    # "meta-llama/llama-3.3-70b-instruct:free",
    # "meta-llama/llama-3.3-70b-instruct:free",
    # "google/gemma-3-27b-it:free",
    # "cognitivecomputations/dolphin-mistral-24b-venice-edition:free",
]
```

```python
def ask_question(model: str) -> str:
    response = requests.post(
        url="https://openrouter.ai/api/v1/chat/completions",
        headers={
            "Authorization": f"Bearer {settings.OPENROUTER_API_KEY}",
            "Content-Type": "application/json",
        },
        data=json.dumps(
            {
                "model": model,
                "messages": [
                    {
                        "role": "user",
                        "content": content,
                    }
                ],
                "reasoning": {"enabled": False},
            }
        ),
    )

    response = response.json()
    try:
        response = response["choices"][0]["message"]
    except KeyError:
        print(response)
        return None
    return response["content"]
```

```python
max_length = max(list(map(len, models)))

for model in models:
    response = ask_question(model)
    print(f"{model:<{max_length}}: {repr(response)}")
```
